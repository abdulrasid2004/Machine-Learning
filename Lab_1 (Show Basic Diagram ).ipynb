```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load datasets
tips = sns.load_dataset("tips")
fmri = sns.load_dataset("fmri")
titanic = sns.load_dataset("titanic")
penguins = sns.load_dataset("penguins")

# Set up the matplotlib figure
plt.figure(figsize=(14, 10))

# 1. Line plot for the "tips" dataset
plt.subplot(2, 2, 1)
sns.lineplot(data=tips, x='total_bill', y='tip', hue='time', marker='o')
plt.title('Line Plot of Tips by Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Tip')

# 2. Bar plot for the "fmri" dataset
plt.subplot(2, 2, 2)
sns.barplot(data=fmri, x='timepoint', y='signal', hue='event')
plt.title('Bar Plot of FMRI Signal by Timepoint')
plt.xlabel('Timepoint')
plt.ylabel('Signal')

# 3. Histogram for the "titanic" dataset
plt.subplot(2, 2, 3)
sns.histplot(data=titanic, x='age', bins=30, kde=True)
plt.title('Histogram of Ages on the Titanic')
plt.xlabel('Age')
plt.ylabel('Frequency')

# 4. Scatter plot for the "penguins" dataset
plt.subplot(2, 2, 4)
sns.scatterplot(data=penguins, x='flipper_length_mm', y='body_mass_g', hue='species')
plt.title('Scatter Plot of Flipper Length vs. Body Mass in Penguins')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Body Mass (g)')

# Adjust layout
plt.tight_layout()
plt.show()
```
